```python
import random as r
import numpy as np
import matplotlib.pyplot as plt
```

```python
def Beneficio(ventas,compras=400):
    if ventas>compras:
        beneficio=ventas*1000+(compras-ventas)*250-compras*750
    else:
        beneficio=compras*1000-compras*750
    return beneficio
```

```python
Beneficio(300)
```

```python
for i in range(10):
    venta = r.random()
    if 0 <= venta < 0.22:
        U_ven = 100
    elif 0.22 <= venta < 0.47:
        U_ven = 200
    elif 0.47 <= venta < 0.58:
        U_ven = 300
    elif 0.58 <= venta < 0.83:
        U_ven = 400
    elif 0.83 <= venta < 0.94:
        U_ven = 500
    else:
        U_ven = 600
    print(U_ven)
```
