```python
import random as r
import numpy as np
import matplotlib.pyplot as plt
```

Paso 1: Definir los beneficios

```python
def Beneficio(ventas,compras=400):
    if ventas<compras:
        beneficio=ventas*1000+(compras-ventas)*250-compras*750
    else:
        beneficio=compras*1000-compras*750
    return beneficio
```

Paso 2: Definir la función de ventas

```python
def Venta():
    venta = r.random()
    if 0 <= venta < 0.22:
        U_ven = 100
    elif 0.22 <= venta < 0.47:
        U_ven = 200
    elif 0.47 <= venta < 0.58:
        U_ven = 300
    elif 0.58 <= venta < 0.83:
        U_ven = 400
    elif 0.83 <= venta < 0.94:
        U_ven = 500
    else:
        U_ven = 600
    return U_ven
```

Paso 3: Ahora se pronosticarán los beneficios promedio

```python
def Ben_prom(n):
  beneficios=[]
  promedio=[]

  for i in range(n):
    v=Venta()
    b=Beneficio(v)
    beneficios.append(b)
    p=np.sum(b)/(i+1)
    promedio.append(p)
  return promedio
```
