```python
import numpy as np
import random as r
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
def f(x):
  return 1/(1+x)
```

```python
f(0.5)
```

```python
def mc1(f,N):
  s = 0
  for i in range (N):
      U = r.random()
      s += f(U)
  return s/N
```

```python
l = []
for i in range (100):
  M = mc1(f,1000)
  l.append(M)
print(np.mean(l))
print(np.var(l))
```

```python
def mc2(f,N):
  s = 0
  for i in range (N):
      U = r.random()
      s += f(1-U)
  return s/N
```

```python
l = []
for i in range (100):
  M = mc2(f,1000)
  l.append(M)
print(np.mean(l))
print(np.var(l))
```

```python
def mc(f,N):
  s = 0
  for i in range (N):
      U = r.random()
      s += (f(U)+f(1-U))/2
  return s/N
```

```python
l = []
for i in range (100):
  M = mc(f,1000)
  l.append(M)
print(np.mean(l))
print(np.var(l))
```

```python
datos = l
n = len(datos)
media = np.mean(datos)
desviacion_estandar = np.std(datos, ddof=1)

# Valor crítico (para 95% de confianza)
alpha = 0.05
valor_z = stats.t.ppf(1 - alpha / 2, df=n-1)

# Margen de error
margen_error = valor_z * (desviacion_estandar / np.sqrt(n))

# Límites
limite_inferior = media - margen_error
limite_superior = media + margen_error

print(f"Intervalo de confianza: ({limite_inferior:.4f}, {limite_superior:.4f})")
```
