```python
import numpy as np
import random as r
import matplotlib.pyplot as plt
```

```python
def GP(lam, T):
  t=0
  S=[]
  I=0
  P=[]
  while T>=t:
    R = r.random()
    t = t-((1/lam)*np.log(R))
    I = I+1
    S.append(t)
    P.append(I)
  return S,P
```

```python
L = GP(4, 10)
```
