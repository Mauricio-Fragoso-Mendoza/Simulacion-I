```python
import numpy as np
import random as r
import matplotlib.pyplot as plt
```

```python
def GP(lam, T):
  t=0
  S=[]
  I=0
  P=[]
  while T>=t:
    R = r.random()
    t = t-((1/lam)*np.log(R))
    I = I+1
    S.append(t)
    P.append(I)
  return S,P
```

```python
L = GP(4, 10)
```

```python
L
```

```python
plt.step(L[0],L[1])
```

```python
L[1][-1] #Última llegada
```

```python
def Adelgazamiento (lam, T):
  t=0
  S=[]
  I=0
  P=[]
  while T>t:
    R1 = r.random()
    t = t-((1/lam)*np.log(R1))
    if 0 <= t <= 5:
      lt=40
    elif 5 < t <= 7:
      lt=60
    else:
      lt=30
    R2 = r.random()
    if R2 <= (lt/lam):
      I=I+1
      S.append(t)
      P.append(I)
  return S,P
```

```python
A = Adelgazamiento(60, 12)
```

```python
plt.step(A[0],A[1])
```
