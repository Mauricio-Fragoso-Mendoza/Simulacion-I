```python
import numpy as np
import random as r
import matplotlib.pyplot as plt
import math
```

```python
def Poisson(lam, n):
  muestra = []
  for _ in range(n):
      R = r.random()
      i = 0
      p = np.exp(-lam) #para P(X=0)
      F = p # F(0)

      while R >= F:
          i += 1
          p = p * lam / i
          F += p

      muestra.append(i)
  return muestra
```

```python
W = Poisson(1,10000)
```

```python
X = Poisson(5,10000)
```

```python
Y = Poisson(10,10000)
```

```python
Z = Poisson(25,10000)
```

```python
plt.hist(W, density=True)
plt.hist(X, density=True)
plt.hist(Y, density=True)
plt.hist(Z, density=True)
```
