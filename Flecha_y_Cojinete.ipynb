**Flecha y Cojinete**

$x_1$ sigue una distribución normal con media $\mu=1.5$ y varianza $\sigma^2=0.0016$; mientras que $x_2$ sigue esa misma distribución con media $\mu=1.48$ y varianza $\sigma^2=0.0009$

```python
import numpy as np
import random as r
import matplotlib.pyplot as plt
```

```python
def fyc(mu,sigma,n):
    m=[]
    for i in range(n):
        suma=0
        for j in range(12):
            u=r.random()
            suma=suma+u
        x=mu+sigma*(suma-6)
        m.append(x)
    return m
```

```python
Y1=fyc(1.5,0.0016,1000)
Y2=fyc(1.48,0.0009,1000)
```

```python
plt.hist(Y1,bins=50)
plt.hist(Y2,bins=50)
plt.show()
```

*a) Probabilidad de interferencia*

Para este caso se tendrá en ambas variables que $x=1$; por lo que los valores de probabilidad serán:
$$P(Z_1) \geq \frac{x-\mu_1}{\sigma_1} \Rightarrow P(Z_1) \geq \frac{1-1.5}{\sigma_1}$$
$$P(Z_2) \geq \frac{x-\mu_2}{\sigma_2} \Rightarrow P(Z_2) \geq \frac{x-1.48}{\sigma_2}$$
