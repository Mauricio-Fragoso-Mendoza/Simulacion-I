**Flecha y Cojinete**

$x_1$ sigue una distribución normal con media $\mu=1.5$ y varianza $\sigma^2=0.0016$; mientras que $x_2$ sigue esa misma distribución con media $\mu=1.48$ y varianza $\sigma^2=0.0009$

```python
import numpy as np
import random as r
import matplotlib.pyplot as plt
```

```python
def fyc(mu,sigma,n):
    m=[]
    for i in range(n):
        suma=0
        for j in range(12):
            u=r.random()
            suma=suma+u
        x=mu+sigma*(suma-6)
        m.append(x)
    return m
```

```python
Y1=fyc(1.5,0.0016,1000)
Y2=fyc(1.48,0.0009,1000)
```

```python
plt.hist(Y1,bins=50)
plt.hist(Y2,bins=50)
plt.show()
```

Para este caso se tiene que el diámetro de la flecha ($X_2$) es inferior al del cojinete ($X_1$), por lo que se tiene interferencia cuando $X2 > X1$

*a) Probabilidad de interferencia*

Siguiendo lo ya antes mencionado, se calculará la diferencia de las medias y la suma de las varianzas con base en los datos de los diámetros de la flecha y el cojinete, es decir, $D = X1 - X2$

Media:
$$\mu_D = \mu_{X_1}-\mu_{X_2}=1.5-1.48 = 0.02$$

Varianza:
$$\sigma_D = \sigma_{X1}+\sigma_{X2} = 0.0016 + 0.0009 = 0.0025$$

Posteriormente, se calculará la probabilidad de intermitencia considerando que $D<0$:
$$Z = \frac{x-\mu_D}{\sigma_D} = \frac{0-0.02}{\sqrt{0.0025}} = \frac{-0.02}{0.05} = -0.4$$

Luego, $$P(D<0) = P(Z<-0.4) = 0.3446$$

*b) Número de simulaciones (con confianza del 95%)*

Para calcular el número de simulaciones se aplicará:
$$n = \frac{Z_{\frac{0.005}{2}}p(1-p)}{E^2}$$

Donde:

$Z_{\frac{0.005}{2}} = Z_{0.025} = 1.96$

$p = P(Z<-0.4) = 0.3446$

Error estimado: $E = 0.01$

```python
z = 1.96
p = 0.3446
E = 0.01

n = (z*p*(1-p))/(E**2)
print('El número de simulaciones necesarias para este proceso serán de ',n)
```

Por lo que se requerirán aproximadamente 4427 simulaciones.
