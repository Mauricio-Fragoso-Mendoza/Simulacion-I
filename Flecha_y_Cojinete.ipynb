**Flecha y Cojinete**

$x_1$ sigue una distribución normal con media $\mu=1.5$ y varianza $\sigma^2=0.0016$; mientras que $x_2$ sigue esa misma distribución con media $\mu=1.48$ y varianza $\sigma^2=0.0009$

```python
import numpy as np
import random as r
import matplotlib.pyplot as plt
```

```python
def fyc1(n):
    m=[]
    for i in range(n):
        suma1=0
        for j in range(12):
            u1=r.random()
            suma1=suma1+u1
        x1=1.50+0.0016*(suma1-6)
    for i in range(n):
        suma2=0
        for j in range(12):
            u2=r.random()
            suma2=suma2+u2
        x2=1.48+0.0009*(suma2-6)
        m.append(x1)
        m.append(x2)
    return m
```

```python
Y=fyc(1000)
```

```python
plt.hist(Y,bins=50)
```
