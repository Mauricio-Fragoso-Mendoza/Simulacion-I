Las hipótesis son:
$$H_0: \mu = \frac{1}{2}$$ $$H_a: \mu \neq \frac{1}{2}$$

El estadístico de prueba es:
$$Z_0 = \frac{\left( \bar{x} - \frac{1}{2} \right) \sqrt{N}}{\sqrt{\frac{1}{12}}}$$

Dónde $\bar{x}$ es el promedio de los datos.
