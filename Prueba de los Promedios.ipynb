```python
import numpy as np
```

```python
def prueba_promedios(datos, alpha):
  n = len(datos)
  if n == 0:
    return 0, 0, 0, False # Return default values for all 4 expected outputs

  promedio = np.mean(datos) # Calculate mean from the input 'datos'

  # Calculations now moved into the function, using its arguments and local 'promedio'
  z_pc = norm.ppf(1 - alpha / 2)
  lim_sup = 0.5 + z_pc * np.sqrt(1 / (12 * n))
  lim_inf = 0.5 - z_pc * np.sqrt(1 / (12 * n))
  z_obt = (promedio - 0.5) / (np.sqrt(1 / (12 * n)))
  valida = lim_inf <= promedio <= lim_sup

  return promedio, z_obt, z_pc, valida
```
