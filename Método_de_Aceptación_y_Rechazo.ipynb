```python
import random as r
import numpy as np
import matplotlib.pyplot as plt
```

```python
def emp(n):
    muestra = []
    for i in range(n):
        u1 = r.random()
        u2 = r.random()
        if u1 > u2:
            muestra.append(u1)
    return muestra
```

```python
L = emp(1000)
```

```python
plt.hist(L, bins=50)
```
