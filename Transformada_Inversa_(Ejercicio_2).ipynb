```python
import random as r
import numpy as np
import matplotlib.pyplot as plt
```

La función $f(x)$ se define como:
$$f(x) = \cases{\frac{1}{b-a}&si $a \leq x \leq b$,\cr 0 &en otro caso}$$

Entonces, $X=(b-a)U+a$

```python
def uniforme(a,b,n):
    l = []
    for i in range(n):
        u = r.random()
        x = (b-a)*u+a
        l.append(x)
    return l
```

```python
R=uniforme(2,5,10000)
```

```python
plt.hist(R, bins=50)
```

```python
def cdf(R):
    n=len(R)
    x=np.sort(R)
    y=np.arange(0,n)/n
    return x,y
```

```python
x_math,y_math=cdf(R)
```

```python
media = np.mean(R)
```

```python
plt.step(x_math,y_math)
plt.axhline(y=0.5, color='red') #La mediana
plt.axvline(x=media, color='green') # La media
```
