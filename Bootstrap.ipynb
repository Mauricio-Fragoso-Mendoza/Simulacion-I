```python
import numpy as np
import pandas as pd
import scipy.stats as stats
```

Los tiempos de entrega de la muestra aleatoria son: $[35,42,38,40,45,37,39,41,44,36,43,40]$

Primero se calcularán las medias de las muestras.

```python
datos = np.array([32, 42, 38, 40, 45, 37, 39, 41, 44, 36, 43, 40])
muestras = 1000
b = []

for i in range(muestras):
    muestra_bs = np.random.choice(datos, size=len(datos), replace=True)
    # Medias de las muestras
    estadistico_bs = np.mean(muestra_bs)
    b.append(estadistico_bs)
```

```python
b
```

Luego, se calcularán la media y la desviación estándar de las 1000 muestras.

```python
print('La media de las muestras es: ',np.mean(b))
print('La desviación estándar de las muestras es: ',np.std(b))
```

Finalmente, se obtendrán los intervalos de confianza del 95% para los percentiles 25% y 97.5%

```python
# Calcular los intervalos de confianza del 95%
LI = np.percentile(b, 2.5)
LS = np.percentile(b, 97.5)

print(f"Intervalo de Confianza: [{LI}, {LS}]")
```
