```python
import numpy as np
import random as r
import matplotlib.pyplot as plt
import scipy.stats as stats
```

Se estimará el resultado de la siguiente integral:
$$I=\int_0^1 e^{x^2}dx$$

Para este caso, se aplicará el Método de Montecarlo para variables antitéticas.

```python
def f(x):
  return np.exp(x**2)
```

```python
f(0)
```

```python
f(1)
```

Se generarán variables $U \sim U(0,1)$, y se generarán 1000 datos.

```python
def va1(f,N):
  s = 0
  for i in range (N):
      U = r.random()
      s += f(U)
  return s/N
```

```python
l = []
for i in range (1000):
  M = va1(f,1000)
  l.append(M)
print(np.mean(l))
print(np.var(l))
```

```python
def va2(f,N):
  s = 0
  for i in range (N):
      U = r.random()
      s += f(1-U)
  return s/N
```

```python
l = []
for i in range (1000):
  M = va2(f,1000)
  l.append(M)
print(np.mean(l))
print(np.var(l))
```
