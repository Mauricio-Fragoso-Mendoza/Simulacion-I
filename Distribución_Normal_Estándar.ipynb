```python
import numpy as np
import random as r
import matplotlib.pyplot as plt
```

```python
def gaussiana(mu,sigma,n):
    muestra=[]
    for i in range(n):
        suma=0
        for j in range(12):
            u=r.random()
            suma=suma+u
        x=mu+sigma*(suma-6)
        muestra.append(x)
    return muestra
```

```python
G=gaussiana(0,1,100000)
```

```python
plt.hist(G, bins=50)
```

```python
def muller(n):
    box=[]
    for i in range(n):
        u1=r.random()
        u2=r.random()
        x=np.sqrt(-2*np.log(u1))*np.cos(2*np.pi*u2)
        y=np.sqrt(-2*np.log(u1))*np.cos(2*np.pi*u2)
        box.append(x)
        box.append(y)
    return box
```

```python
M=muller(10)
```

```python
plt.hist(M,bins=50)
```
