```python
import numpy as np
import random as r
import math as m
import matplotlib.pyplot as plt
```

Los elementos a destacar son:

$t$: Tiempo

$N_A$: Número de llegadas

$N_D$: Número de salidas

Eventos:

$t_A$: Tiempo de la siguiente llegada

$t_D$: Tiempo de la siguiente salida

Queremos:

a) Tiempo promedio de un cliente en el sistema

b) Tiempo en que sale el último cliente

Generar $T_0 \rightarrow U \backsim U(0,1)$
$$t=t-\frac{1}{\lambda}\log U$$
$$t_A=T_0$$
$$t_D=\infty$$

```python
def SLE(lam, T):
  t=0
  S=[]
  NA=0
  A=[]
  ND=0
  D=[]
  while T>=t:
    T0=r.random()
    t = t-((1/lam)*np.log(T0))
    n=0
    tA=T0
    tD=m.inf
    if tA<=tD and tA<=T:
      t=tA
      NA=NA+1
      n=n+1
      Ti=r.random()
      tA=Ti
      if n==1:
        Y=r.random()
        tD=t+Y
      A.append(t)
    elif tD<tA and tD<=T:
      t=tD
      ND=ND+1
      n=n-1
      if n==0:
        tD=m.inf
      else:
        Y=r.random()
        tD=t+Y
      D.append(t)
    elif min(tA,tD)>T and n>0:
      t=tD
      ND=ND+1
      n=n-1
      if n>0:
        Y=r.random()
        tD=t+Y
      D.append(t)
    elif min(tA,tD)>T and n==0:
      Tp=max(t-T,0)
  return S,A,D,Tp
```

```python
Z=SLE(10,12)
```

```python
plt.step(Z[0],Z[1])
```
