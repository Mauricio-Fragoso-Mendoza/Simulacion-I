```python
import numpy as np
import random as r
import matplotlib.pyplot as plt
```

```python
def f(n):
  L=[]
  for i in range(n):
     R=r.random()
     if 0<=R<=0.2:
       x=1
     elif 0.2<R<=0.35:
       x=2
     elif 0.35<R<=0.6:
       x=3
     else:
       x=4
     L.append(x)
  return L
```

```python
M=f(1000)
```

```python
plt.hist(M,density=True)
```
