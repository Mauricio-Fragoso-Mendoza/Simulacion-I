```python
import numpy as np
import random as r
import matplotlib.pyplot as plt
```

```python
def f(x):
    return np.sqrt(np.arctan(x))
```

```python
xx = np.linspace(0,1,1000)
plt.plot(xx,f(xx))
```

```python
f(1)
```

```python
def mcc(a=0, b=1, N=10):
    I = 0
    for i in range(N):
        x = r.random()
        I += f(x)
    return I/N
```

```python
mcc(0,1,1000)
```

```python
def experimento(M=600):
    L=[]
    for i in range(M):
        I=mcc(0,1,i+1)
        L.append(I)
    return L
```

```python
E=experimento(10)
```

```python
plt.plot(E)
```

```python
E=experimento(100)
```

```python
plt.plot(E)
```

```python
E=experimento(1000)
```

```python
plt.plot(E)
```

```python
np.mean(E)
```

```python
def simulacion(F=30):
    Li=[]
    for i in range(F):
        e=experimento(600)
        Li.append(e)
    return Li
```

```python
s=simulacion(100)
```
